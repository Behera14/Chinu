```python
# Non-Recursive (Iterative) Fibonacci
def fibonacci(n):
    a, b = 0, 1
    steps = 0
    print("Fibonacci Series:", end=" ")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
        steps += 1
    print("\nStep Count:", steps)

n = int(input("Enter number of terms: "))
fibonacci(n)
```

```python
# Recursive Fibonacci with Step Counter
steps = 0  # global counter

def fibonacci_recursive(n):
    global steps
    steps += 1
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

n = int(input("Enter number of terms: "))
print("Fibonacci Series (Recursive):", end=" ")
for i in range(n):
    print(fibonacci_recursive(i), end=" ")

print("\nStep Count:", steps)
```

```python
## 2. Implement job sequencing with deadlines using a greedy method. 
```

```python
class Job: 
    def __init__(self, id, deadline, profit): 
        self.id = id 
        self.deadline = deadline 
        self.profit = profit 
 
def job_sequencing(jobs): 
    # Sort by decreasing profit 
    jobs.sort(key=lambda x: x.profit, reverse=True) 
    max_deadline = max(job.deadline for job in jobs) 
    slot = [-1] * (max_deadline + 1) 
    total_profit = 0 
 
    for job in jobs: 
        for j in range(job.deadline, 0, -1): 
            if slot[j] == -1: 
                slot[j] = job.id 
                total_profit += job.profit 
                break 
 
    print("Job Sequence:", [j for j in slot if j != -1]) 
    print("Total Profit:", total_profit) 
 
n = int(input("Enter number of jobs: ")) 
jobs = [] 
for i in range(n): 
    job_id = input(f"Enter Job ID {i+1}: ") 
    deadline = int(input("Enter Deadline: ")) 
    profit = int(input("Enter Profit: ")) 
    jobs.append(Job(job_id, deadline, profit)) 
 
job_sequencing(jobs)
```

```python
 #2. write a program to implement huffman encoding using a greedy stratergy 
```

```python
import heapq

# Create a class to represent a Huffman tree node
class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    # Define comparison for priority queue (min-heap)
    def __lt__(self, other):
        return self.freq < other.freq


# Function to build Huffman Tree
def build_huffman_tree(char_freq):
    heap = [Node(char, freq) for char, freq in char_freq.items()]
    heapq.heapify(heap)  # Convert list into a min-heap

    while len(heap) > 1:
        # Pick two smallest nodes
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)

        # Merge them into a new node
        merged = Node(None, left.freq + right.freq)
        merged.left = left
        merged.right = right

        # Push the new node back into heap
        heapq.heappush(heap, merged)

    # Root of Huffman Tree
    return heap[0]


# Generate Huffman Codes by traversing the tree
def generate_codes(node, code="", huffman_codes={}):
    if node is None:
        return

    # If leaf node, store its code
    if node.char is not None:
        huffman_codes[node.char] = code
        return

    generate_codes(node.left, code + "0", huffman_codes)
    generate_codes(node.right, code + "1", huffman_codes)
    return huffman_codes


# Main program
if __name__ == "__main__":
    text = input("Enter text to encode: ")

    # Step 1: Calculate frequency of each character
    freq = {}
    for ch in text:
        freq[ch] = freq.get(ch, 0) + 1
    print("\nCharacter Frequencies:", freq)

    # Step 2: Build Huffman Tree
    root = build_huffman_tree(freq)

    # Step 3: Generate Huffman Codes
    huffman_codes = generate_codes(root)
    print("\nHuffman Codes:")
    for ch, code in huffman_codes.items():
        print(ch, ":", code)

    # Step 4: Encode the text
    encoded_text = ''.join(huffman_codes[ch] for ch in text)
    print("\nEncoded Text:", encoded_text)

    # Step 5: Display compression information
    print("\nOriginal size (bits):", len(text) * 8)
    print("Encoded size (bits):", len(encoded_text))
```

```python
#3. Write a program to solve a fractional Knapsack problem using a greedy method. 
```

```python
def fractional_knapsack(weights, profits, capacity):
    n = len(weights)
    ratio = [(profits[i] / weights[i], i) for i in range(n)]
    ratio.sort(reverse=True)

    total_profit = 0.0

    for r, i in ratio:
        if capacity >= weights[i]:
            capacity -= weights[i]
            total_profit += profits[i]
        else:
            total_profit += r * capacity
            break

    print("Maximum Profit:", total_profit)


# --- Main Program ---
n = int(input("Enter number of items: "))
weights = list(map(float, input("Enter weights: ").split()))
profits = list(map(float, input("Enter profits: ").split()))
capacity = float(input("Enter capacity of knapsack: "))

fractional_knapsack(weights, profits, capacity)
```

```python
#4. Write a program to solve a 0-1 Knapsack problem using dynamic programming or branch and bound strategy
```

```python
def knapsack_01(weights, profits, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(profits[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    print("Maximum Profit:", dp[n][capacity])


n = int(input("Enter number of items: "))
weights = list(map(int, input("Enter weights: ").split()))
profits = list(map(int, input("Enter profits: ").split()))
capacity = int(input("Enter capacity of knapsack: "))

knapsack_01(weights, profits, capacity)
```

```python
#5. Design 8-Queens matrix having first Queen placed. Use backtracking to place remaining Queens to generate the final 8
#queen’s matrix. 
```

```python
N = 8

def print_solution(board):
    for row in board:
        print(" ".join("Q" if x == 1 else "." for x in row))
    print()

def is_safe(board, row, col):
    # Check this row on the left
    for i in range(col):
        if board[row][i] == 1:
            return False

    # Check upper-left diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check lower-left diagonal
    for i, j in zip(range(row, N), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    return True

def solve(board, col):
    # Base case: If all queens are placed, print the board
    if col >= N:
        print_solution(board)
        return True

    # Try placing queen in all rows of this column
    for i in range(N):
        if is_safe(board, i, col):
            board[i][col] = 1

            # Recur to place the rest of the queens
            if solve(board, col + 1):
                return True

            # Backtrack
            board[i][col] = 0

    return False

# Initialize board
board = [[0] * N for _ in range(N)]

# Take input for first queen position
first_row = int(input("Enter row position for first queen (0-7): "))
board[first_row][0] = 1

# Solve from next column
solve(board, 1)
```

```python
import java.util.Scanner;

public class NQueens {
    static final int N = 8;

    // Print the chessboard solution
    static void printSolution(int[][] board) {
        for (int i = 0; i < N; i++) {
            for (int j = 0; j < N; j++) {
                System.out.print(board[i][j] == 1 ? "Q " : ". ");
            }
            System.out.println();
        }
        System.out.println();
    }

    // Check if it's safe to place a queen at board[row][col]
    static boolean isSafe(int[][] board, int row, int col) {
        // Check this row on the left
        for (int i = 0; i < col; i++) {
            if (board[row][i] == 1)
                return false;
        }

        // Check upper-left diagonal
        for (int i = row, j = col; i >= 0 && j >= 0; i--, j--) {
            if (board[i][j] == 1)
                return false;
        }

        // Check lower-left diagonal
        for (int i = row, j = col; i < N && j >= 0; i++, j--) {
            if (board[i][j] == 1)
                return false;
        }

        return true;
    }

    // Solve the N-Queens problem recursively
    static boolean solve(int[][] board, int col) {
        // Base case: all queens are placed
        if (col >= N) {
            printSolution(board);
            return true;
        }

        // Try placing a queen in each row of this column
        for (int i = 0; i < N; i++) {
            if (isSafe(board, i, col)) {
                board[i][col] = 1; // Place queen

                // Recur to place the rest of the queens
                if (solve(board, col + 1))
                    return true;

                // Backtrack
                board[i][col] = 0;
            }
        }

        return false;
    }

    public static void main(String[] args) {
        int[][] board = new int[N][N];
        Scanner sc = new Scanner(System.in);

        System.out.print("Enter row position for first queen (0-7): ");
        int firstRow = sc.nextInt();

        // Place the first queen in the first column
        board[firstRow][0] = 1;

        // Solve for remaining columns
        if (!solve(board, 1))
            System.out.println("No solution found!");

        sc.close();
    }
}
```
